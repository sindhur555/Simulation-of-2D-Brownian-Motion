```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd
from random import randint
from matplotlib.animation import FuncAnimation
import sys
import ffmpeg
```

```python
x=[]
y=[]

x.append(np.random.normal(loc=0, scale = 1, size=1))
y.append(np.random.normal(loc=0, scale = 1, size=1))
fig,ax = plt.subplots()

ax.clear()

def animate(i):
    x.append(x[i-1]+np.random.normal(loc=0, scale = 1, size=1))
    y.append(y[i-1]+np.random.normal(loc=0, scale = 1, size=1))
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-plot")
    ax.set_title("2D Brownian Motion")
    ax.clear()
    ax.plot(x, y)

    # run the animation
ani = FuncAnimation(fig, animate, frames=30, interval=500, repeat=False)

#plt.show()
    
from matplotlib import rc

# equivalent to rcParams['animation.html'] = 'html5'
rc('animation', html='html5')
ani
```
